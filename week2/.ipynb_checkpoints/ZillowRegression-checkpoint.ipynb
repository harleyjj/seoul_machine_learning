# Fire up graphlab create

```python
import graphlab
```

# Load our cleaned features

Dataset is from week 1, with a few more steps taken (now part of Jupyter Notebook in week1 folder of Github repository)

```python
features_plus_error = graphlab.SFrame('features_with_error')
```

```python
features_plus_error
```

# Exploring the data for housing sales 

The magnitude of error seems to be higher when lot size is low.  

```python
graphlab.canvas.set_target('ipynb')
features_plus_error.show(view="Scatter Plot", x="lotsizesquarefeet", y="abslogerror")
```

# Create a simple regression model of sqft_living to price

Split data into training and testing.  
We use seed=0 so that everyone running this notebook gets the same results.  In practice, you may set a random seed (or let GraphLab Create pick a random seed for you).  

```python
train_data,test_data = features_plus_error.random_split(.8,seed=0)
```

## Build the regression model using only sqft_living as a feature

```python
lot_model = graphlab.linear_regression.create(train_data, target='abslogerror', features=['lotsizesquarefeet'], validation_set=None)
```

# Evaluate the simple model

```python
print test_data['abslogerror'].mean()
```

```python
print lot_model.evaluate(test_data)
```

RMSE of about 0.152

# Let's show what our predictions look like

Matplotlib is a Python plotting library that is also useful for plotting.  You can install it with:

'pip install matplotlib'

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(test_data['lotsizesquarefeet'],test_data['abslogerror'],'.',
        test_data['lotsizesquarefeet'],lot_model.predict(test_data),'-')
```

Above:  blue dots are original data, green line is the prediction from the simple regression.

Below: we can view the learned regression coefficients. 

```python
lot_model.get('coefficients')
```

# Explore other features in the data

To build a more elaborate model, we will explore using more features.

```python
my_features = ['lotsizesquarefeet', 'roomcnt', 'taxvaluedollarcnt', 'taxamount']
```

```python
features_plus_error[my_features].show()
```

0 rooms seems to have the highest errors.

# Build a regression model with more features

```python
my_features_model = graphlab.linear_regression.create(train_data,target='abslogerror',features=my_features,validation_set=None)
```

```python
print my_features
```

## Comparing the results of the simple model with adding more features

```python
print lot_model.evaluate(test_data)
print my_features_model.evaluate(test_data)
```

 The RMSE goes down from 0.1518 to 0.1516 with more features.

```python
plt.plot(test_data['lotsizesquarefeet'],test_data['abslogerror'],'.',
        test_data['lotsizesquarefeet'],lot_model.predict(test_data),'-')
```

```python
plt.plot(test_data['lotsizesquarefeet'],test_data['abslogerror'],'.',
        test_data['lotsizesquarefeet'],my_features_model.predict(test_data),'-')
```
